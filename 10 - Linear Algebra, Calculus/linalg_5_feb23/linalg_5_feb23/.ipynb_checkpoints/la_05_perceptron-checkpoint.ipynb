```python
def show(w, w0, x, y):
    line_x = np.linspace(-5, 10)
    line_y = (-w[0]*line_x - w0)/w[1] 
    plt.plot(line_x, line_y)                        # show h-plane
    plt.scatter(x[:, 0], x[:, 1], c=y)
    plt.show()
```

```python
w = np.random.normal(size=(2,))
w0 = np.random.normal()

show(w, w0, x, y)
```

```python
print(x.shape, y.shape)
```

```python
def algorithm(x, y, w, w0, num_iterations=100):
    for iteration in range(num_iterations):
        miss_cls = 0
        for point in range(len(x)):
            hs = np.sign(np.dot(w, x[point]) + w0)
            if hs != np.sign(y[point]):
                miss_cls += 1
                w = w+x[point]*np.sign(y[point]) # courtesy Sintu
                w0 += np.sign(y[point])
                break # one misclassification at a time
        if miss_cls == 0:
            show(w, w0, x, y)
            return w, w0
    show(w, w0, x, y)
    return w, w0
                
```

```python
w = np.random.normal(size=(2,))
w0 = np.random.normal()
w, w0
```

```python
show(w, w0, x, y)
```

```python
w, w0 = algorithm(x, y, w, w0, num_iterations=1)
```

```python
new_point = np.array([2, 15])
np.sign(np.dot(w, new_point) + w0)
```

```python
# We use all points that are misclassified
def algorithm2(x, y, w, w0):
    for iteration in range(5000):
        miss_cls = 0
        sum_miss_cls_vect = 0
        sum_bias = 0
        for point in range(len(x)):
            hs = np.sign(np.dot(w, x[point]) + w0) # half space
            if hs != np.sign(y[point]):
                miss_cls += 1
                sum_miss_cls_vect += x[point]*np.sign(y[point])
                sum_bias += np.sign(y[point])

        if ((miss_cls / len(x)) < 0.02):
            show(w, w0, x, y)
            print('N_iter', iteration, '| 98% accuracy!!')
            break
        
        w = w + sum_miss_cls_vect / miss_cls
        w0 = w0 + sum_bias / miss_cls
        
        
        if iteration % 500 == 0:
            show(w, w0, x, y)
            print('N_iter', iteration, '| Accuracy:', 1 - round(miss_cls / len(x), 2))
    return w, w0
```

```python
w, w0 = algorithm2(x, y, w, w0)
```
